```python
f = open('/data/QB_data.csv')
csv_f = csv.reader(f)

QBs = []
QB_ids = []
for row in csv_f:
    name = row[0].split("\\")
    QBs.append(name)

for QB in QBs[1:]:
    QB_ids.append(QB[1])

QB_ids.remove('JohnAl20')
QBdataset = pd.DataFrame()
for QB_id in QB_ids:
    print(QB_id)
    QBdataset = pd.concat([QBdataset, Player(QB_id).dataframe])
```

```python
for col in QBdataset:
    QBdataset[col] = pd.Categorical(QBdataset[col])
    command = "QBdataset[col] = QBdataset." + col + ".cat.codes"
    exec(command)
        
QBdataset['HOF'] = QBdataset['HOF'].astype('bool')
QBdataset.fillna(0)
FIELDS_TO_DROP = ['name', 'birth_date', 'height', 'weight', 'position', 'team_abbreviation', 'HOF',]
```

```python
X = QBdataset.drop(FIELDS_TO_DROP, 1).drop_duplicates()
y = QBdataset['HOF'].values
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = Sequential()
model.add(Dense(104, input_dim=104, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model.fit(X_train, y_train,
          epochs=20,
          batch_size=128)
classification = model.evaluate(X_test, y_test, batch_size=128)
```
